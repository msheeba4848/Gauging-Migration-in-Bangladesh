```python
import xml.etree.ElementTree as ET

xml_data = """<?xml version="1.0" encoding="UTF-8" standalone="yes"?>
<a:clrScheme xmlns:a="http://schemas.openxmlformats.org/drawingml/2006/main" name="PGO 013 - Urban Sunset">
    <a:dk1><a:srgbClr val="2D3740"/></a:dk1>
    <a:lt1><a:srgbClr val="FFFFFF"/></a:lt1>
    <a:dk2><a:srgbClr val="635B4E"/></a:dk2>
    <a:lt2><a:srgbClr val="E7E6E6"/></a:lt2>
    <a:accent1><a:srgbClr val="2D3740"/></a:accent1>
    <a:accent2><a:srgbClr val="E7943F"/></a:accent2>
    <a:accent3><a:srgbClr val="FE5C4F"/></a:accent3>
    <a:accent4><a:srgbClr val="635B4E"/></a:accent4>
    <a:accent5><a:srgbClr val="B09471"/></a:accent5>
    <a:accent6><a:srgbClr val="F7D290"/></a:accent6>
    <a:hlink><a:srgbClr val="AC5448"/></a:hlink>
    <a:folHlink><a:srgbClr val="FE5C4F"/></a:folHlink>
</a:clrScheme>
"""

# Define namespace
namespace = {'a': 'http://schemas.openxmlformats.org/drawingml/2006/main'}

# Parse XML
root = ET.fromstring(xml_data)

# Extract hex colors
hex_colors = [child.find('a:srgbClr', namespace).attrib['val'] for child in root]

# Output hex colors
print("Hex Colors:")
for color in hex_colors:
    print(color)
```

```python
import matplotlib.pyplot as plt

# Hex colors
hex_colors = [
    "#2D3740", "#FFFFFF", "#635B4E", "#E7E6E6",
    "#2D3740", "#E7943F", "#FE5C4F", "#635B4E",
    "#B09471", "#F7D290", "#AC5448", "#FE5C4F"
]

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot each color as a square
for i, color in enumerate(hex_colors):
    ax.add_patch(plt.Rectangle((i % 4, i // 4), 1, 1, color=color))

    # Add text with hex value
    ax.text((i % 4) + 0.5, (i // 4) + 0.5, color, color='black', ha='center', va='center')

# Set axis limits and labels
ax.set_xlim(0, 4)
ax.set_ylim(0, 3)
ax.set_xticks(range(5))
ax.set_yticks(range(4))
ax.set_xticklabels([])
ax.set_yticklabels([])

# Remove axis
ax.axis('off')

# Show plot
plt.show()
```
