```python
import pandas as pd
```

```python
df1 = pd.read_csv('ICPSR_38846/DS0001/38846-0001-Data.tsv', sep='\t')

print(df1.shape)
print(df1.columns)
print(df1.head())
print(df1.count())
```

```python
import pandas as pd

# Set the display option to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Calculate the proportion of missing values in each column
missing_proportion = (df1.isnull().mean() * 100).round(2)

# Create a DataFrame to store the result
missing_df = pd.DataFrame({'Column': missing_proportion.index, 'Missing Percentage': missing_proportion.values})

# Display the DataFrame
print(missing_df)
```
