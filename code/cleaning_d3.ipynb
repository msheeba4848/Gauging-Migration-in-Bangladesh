```python
import pandas as pd

# Set the display option to show all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Calculate the proportion of missing values in each column
missing_proportion = (df3.isnull().mean() * 100).round(2)

# Create a DataFrame to store the result
missing_df = pd.DataFrame({'Column': missing_proportion.index, 'Missing Percentage': missing_proportion.values})

# Display the DataFrame
print(missing_df)
```

```python
subset3 = df3
# replace the spaces with NA to review NA values with numpy and pandas functions
subset3.replace(' ', np.nan, inplace=True)
counts = subset3.count()
columns_to_drop = counts[counts < 1000].index

# columns to drop
subset3.drop(columns=columns_to_drop, inplace=True)

print(subset3.head())
print(subset3.shape)
```

```python
subset3 = df3
subset3 = subset3.apply(pd.to_numeric, errors='coerce')
subset3.replace(' ', np.nan, inplace=True)
mean_cols = subset3.mean()
subset3 = subset3.loc[:, mean_cols != 0]

print(subset3.shape)
```

```python
counts = subset3.count()
print(counts[:10])
print(counts[10:20])
print(counts[20:30])
print(counts[40:50])
print(counts[50:60])
print(counts[60:70])
print(counts[70:80])
print(counts[80:90])
print(counts[90:100])
print(counts[100:110])
print(counts[110:120])
print(counts[120:130])
print(counts[130:140])
print(counts[140:150])
print(counts[150:154])
```
